```python
import re

def normalize_sll_crime_head(s):
    if pd.isna(s):
        return s

    s = s.strip()

    # Fix encoding issues
    s = s.replace("�", "-")

    # Normalize dashes
    s = s.replace("–", "-").replace("—", "-")

    # Remove multiple spaces
    s = re.sub(r"\s+", " ", s)

    # Standardize case for matching (keep original text)
    return s
```

```python
def categorize_sll_crime(crime):
    c = crime.lower()

    # ───────── Crime Against Women ─────────
    if any(k in c for k in [
        "dowry", "domestic violence",
        "indecent representation of women",
        "immoral traffic",
        "crime against women",
        "transgender persons"
    ]):
        return "Crime Against Women"

    # ───────── Crime Against Children ─────────
    if any(k in c for k in [
        "children", "juvenile justice",
        "child marriage", "child labour",
        "pre-natal diagnostic",
        "medical termination of pregnancy",
        "pocso"
    ]):
        return "Crime Against Children"

    # ───────── SC/ST Atrocity Crimes ─────────
    if any(k in c for k in [
        "sc/st", "scheduled castes",
        "scheduled tribes",
        "civil rights act",
        "against sc", "against st"
    ]):
        return "SC/ST Atrocity Crimes"

    # ───────── Public Order & State Security ─────────
    if any(k in c for k in [
        "unlawful activities",
        "official secrets",
        "offences against state",
        "national honour",
        "representation of the people",
        "organised crimes",
        "macoca"
    ]):
        return "Public Order & State Security Crimes"

    # ───────── Arms & Explosives ─────────
    if any(k in c for k in [
        "arms act", "fire arms",
        "explosive", "arms/explosive"
    ]):
        return "Arms & Explosives Crimes"

    # ───────── Cyber / IP Crimes ─────────
    if any(k in c for k in [
        "information technology",
        "copyright", "trade marks",
        "intellectual property"
    ]):
        return "Cyber / Intellectual Property Crimes"

    # ───────── Economic / Financial Crimes ─────────
    if any(k in c for k in [
        "lotteries", "chit funds",
        "negotiable instruments",
        "benami", "mines and minerals",
        "prevention of corruption",
        "finance & economic"
    ]):
        return "Economic / Financial Crimes"

    # ───────── Narcotics & Liquor ─────────
    if any(k in c for k in [
        "narcotic drugs",
        "psychotropic",
        "liquor", "excise",
        "prohibition act",
        "possession of drugs"
    ]):
        return "Narcotics & Liquor Crimes"

    # ───────── Environment & Pollution ─────────
    if any(k in c for k in [
        "forest", "wildlife",
        "environment",
        "pollution", "air & water",
        "tobacco", "noise",
        "green tribunal"
    ]):
        return "Environment & Pollution Crimes"

    # ───────── Immigration / Foreigner ─────────
    if any(k in c for k in [
        "foreigners", "passport",
        "emigration", "citizenship"
    ]):
        return "Immigration & Foreigner Crimes"

    # ───────── Transport & Infrastructure ─────────
    if any(k in c for k in [
        "railways", "railway property",
        "telegraph", "cinematograph",
        "cable television",
        "motor vehicle"
    ]):
        return "Transport & Infrastructure Crimes"

    # ───────── Food / Drugs / Commodities ─────────
    if any(k in c for k in [
        "essential commodities",
        "food adulteration",
        "food safety"
    ]):
        return "Food, Drugs & Essential Commodities Crimes"

    # ───────── Regulatory / Local Acts ─────────
    if any(k in c for k in [
        "gambling", "electricity",
        "antiques", "bonded labour",
        "defacement",
        "state local acts",
        "city/town police acts",
        "other regulatory"
    ]):
        return "Regulatory / Local Law Crimes"

    # ───────── Catch-All ─────────
    if any(k in c for k in [
        "other sll", "total cognizable"
    ]):
        return "Miscellaneous / Other SLL Crimes"

    return "Miscellaneous / Other SLL Crimes"
```

```python
rename_maps = {
    2017: sll_2017_rename,
    2018: sll_2018_rename,
    2019: sll_2019_rename,
    2020: sll_2020_rename,
    2021: sll_2021_rename
}

sll_dfs = {
    2017: df17,
    2018: df18,
    2019: df19,
    2020: df20,
    2021: df21
}

for year in sll_dfs:
    df = sll_dfs[year]
    df = df.rename(columns=rename_maps[year])
    df["crime_head"] = df["crime_head"].apply(normalize_sll_crime_head)
    df["crime_type"] = df["crime_head"].apply(categorize_sll_crime)
    
    sll_dfs[year] = df
```
